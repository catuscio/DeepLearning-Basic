```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
import numpy as np
import pandas as pd

#seed settings
np.random.seed(3)
tf.random.set_seed(3)

#dataset settings
dataset = df.values
X = dataset[:,2:6].astype(float)
Y_obj = dataset[:,6]

#str to num
e = LabelEncoder()    # sklearn 라이브러리 
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y_encoded = tf.keras.utils.to_categorical(Y)

# 모델 설정
model = Sequential() 
model.add(Dense(16,  input_dim=4, activation='relu'))   # 은닉층
model.add(Dense(3, activation='softmax'))   # 출력층층

# 모델 컴파일
model.compile(loss='categorical_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])

# 모델 학습
model.fit(X, Y_encoded, epochs=300, batch_size=5)

# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y_encoded)[1]))
```
