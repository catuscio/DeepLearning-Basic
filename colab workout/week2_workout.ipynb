```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import statsmodels.api as statm
import statsmodels.formula.api as statfa

x = np.array([[3, 6], [6, 4], [9, 5], [12, 2]])
y = np.array([70, 75, 80, 90])

x1 = [i[0] for i in x]
x2 = [i[1] for i in x]

model = Sequential()    #sequential 
model.add(Dense(1, input_dim=2, activation = 'linear'))   #dimension = 2, linear graph
model.compile(optimizer='sgd', loss='mse')  #set optimizing & loss calculation method(gradient discent/mse)
model.fit(x, y, epochs=2000)

hour = 8
dope = 4
prediction = model.predict([[hour, dope]])

print("hour = %d, lesson = %d, estimated score = %.2f" %(hour, dope, prediction))
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('study hours')
ax.set_ylabel('private_class')
ax.set_zlabel('estimated score')
ax.dist = 11
ax.scatter(x1, x2, y)
plt.show()
```
